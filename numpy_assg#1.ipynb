```python
def plti(im, h = 4, **kwargs):
    """
    Helper function to plot image.
    """
    y = im.shape[0]
    x = im.shape[1]
    w = (y/x) * h
    plt.figure(figsize=(w,h))
    plt.imshow(im, interpolation="none", **kwargs)
```

```python
import numpy as np
import os
import matplotlib.pylab as plt
from PIL import Image

size = (200,200)
imagesDir = os.getcwd()+'/images'
myarr = np.zeros([20,200,200,3], dtype=int)

j = 0
for f in os.listdir(imagesDir):
    link = "images/" + f
    i = Image.open(link)
    fn, fext = os.path.splitext(f)
    fext = ".png"
    resizedImage = i.resize((200,200))
    resizedImage.save("200/{}_200{}".format(fn,fext))
    myarr[j] = resizedImage
    j += 1
  
for each in range(20):
    plti(myarr[each])
```
